```python
import pandas as pd
import numpy as np

X = [85, 60, 75, 90, 70]
Y = [80, 70, 65, 95, 60]

X = np.array([85, 60, 75, 90, 70])
Y = np.array([80, 70, 65, 95, 60])

rank_X = X.argsort().argsort()
rank_Y = Y.argsort().argsort()

print("Rank X:", rank_X)
print("Rank Y:", rank_Y)
```

```python
rank_X = X.argsort().argsort() + 1
rank_Y = Y.argsort().argsort() + 1

print("Rank X:", rank_X)
print("Rank Y:", rank_Y)
```

```python
# Compute difference between Ranks

d = rank_X - rank_Y
d_square =d**2
print("d:", d)
print("d_square:", d_square)
```

```python

n = len(X)
sum_d2 = np.sum(d_square)

r_s = 1 - (6 * sum_d2) / (n * (n**2 - 1))
print("Spearman Rank Correlation =", r_s)
```
