```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
# Demonstrate the sampling distribution of mean and  variance
# Step 1:Create a Population
population = np.random.normal(loc = 50, scale =10, size = 10000)

#Step 2: take many random samples
sample_size = 30
num_samples = 1000

sample_means= []

for i in range(num_samples):
     sample = np.random.choice(population, sample_size)
     sample_means.append(np.mean(sample))

#Step 3: plot sampling distribution of mean
plt.hist(sample_means, bins = 30)
plt.title("Sampling Distribution of Mean")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()



```
