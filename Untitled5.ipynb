```python
import matplotlib.pyplot as plt
import numpy as np

# Sample Data
subjects = ['Math', 'Science', 'English','History']
marks = [85, 90, 78, 88]

months = ['Jan', 'Feb', 'Mar', 'Apr']
sales = [100, 150, 200, 180]

#Bar  Chart
plt.figure()
plt.bar(subjects, marks)
plt.title('Bar Chart -  Student Marks')
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.show()

# Pie Chart

plt.figure()
plt.pie(marks, labels=subjects, autopct='%1.1f%%')
plt.title('Pie Chart - Student Marks')
plt.show()

#Line Chart

plt.figure()
plt.plot(months, sales, marker='o', linestyle='-')
plt.title('Line Graph - Sales Data')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.show()

```
