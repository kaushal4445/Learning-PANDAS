```python
import pandas as pd

# Define symptoms and associated diseases
disease_database = {
    "fever": ["Flu", "Malaria", "Dengue"],
    "headache": ["Migraine", "Flu", "Cold"],
    "cough": ["Cold", "Flu", "Tuberculosis"],
    "chest pain": ["Heart Disease", "Pneumonia"],
    "vomiting": ["Food Poisoning", "Malaria"],
    "rash": ["Allergy", "Measles", "Dengue"]
}

# Ask user for symptoms
print("Enter your symptoms from the following list:")
print(", ".join(disease_database.keys()))

user_symptoms = input("Enter symptoms separated by commas: ").lower().split(",")

# Clean whitespace
user_symptoms = [symptom.strip() for symptom in user_symptoms]

# Collect possible diseases
possible_diseases = []

for symptom in user_symptoms:
    if symptom in disease_database:
        possible_diseases.extend(disease_database[symptom])
    else:
        print(f"Unknown symptom: {symptom}")

# Count disease frequency and suggest most probable
if possible_diseases:
    from collections import Counter
    disease_count = Counter(possible_diseases)
    most_probable = disease_count.most_common(1)[0][0]
    print("\nBased on your symptoms, you may have:", most_probable)
else:
    print("\nNo disease could be diagnosed with the given symptoms.")
```
