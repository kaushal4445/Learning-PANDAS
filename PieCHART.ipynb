```python
import matplotlib.pyplot as plt
Subjects = ["DBMS", "DSA", "OS"]
Marks = [60, 70, 80]

plt.pie(Marks, labels=Subjects, autopct="%1.3f%%")
plt.title("Pie Chart")
plt.show()
```
