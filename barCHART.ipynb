```python
import matplotlib.pyplot as plt
Subjects = ["DBMS", "DSA", "OS"]
Marks = [60, 70, 80]

plt.bar(Subjects, Marks)
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title(" BAR CHART")
plt.show()
```
